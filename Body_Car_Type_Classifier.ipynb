### Body car type classifier with data augmentation and transfer learning

```python
import pandas as pd                                     # Data analysis and manipultion tool
import numpy as np                                      # Fundamental package for linear algebra and multidimensional arrays
import tensorflow as tf                                 # Deep Learning Tool
import os                                               # OS module in Python provides a way of using operating system dependent functionality
import cv2                                              # Library for image processing
from sklearn.model_selection import train_test_split    # For splitting the data into train and validation set
from PIL import Image
```

```python
labels = pd.read_csv('...\standford_cars/Training_set.csv')   # loading the labels

Sedan_files =[]
for fname, lab in zip(labels['filename'], labels['label']):
  if lab =='Sedan':
    Sedan_files.append(fname)
Sedan_files, len(Sedan_files) 

Convertible_files =[]
for fname, lab in zip(labels['filename'], labels['label']):
  if lab =='Convertible':
    Convertible_files.append(fname)
Convertible_files, len(Convertible_files) 

SUV_files =[]
for fname, lab in zip(labels['filename'], labels['label']):
  if lab =='SUV':
    SUV_files.append(fname)
SUV_files, len(SUV_files) 

Coupe_files =[]
for fname, lab in zip(labels['filename'], labels['label']):
  if lab =='Coupe':
    Coupe_files.append(fname)
Coupe_files, len(Coupe_files) 

Cab_files =[]
for fname, lab in zip(labels['filename'], labels['label']):
  if lab =='Cab':
    Cab_files.append(fname)
Cab_files, len(Cab_files) 

Hatchback_files =[]
for fname, lab in zip(labels['filename'], labels['label']):
  if lab =='Hatchback':
    Hatchback_files.append(fname)
Hatchback_files, len(Hatchback_files) 

Minivan_files =[]
for fname, lab in zip(labels['filename'], labels['label']):
  if lab =='Minivan':
    Minivan_files.append(fname)
Minivan_files, len(Minivan_files) 

Wagon_files =[]
for fname, lab in zip(labels['filename'], labels['label']):
  if lab =='Wagon':
    Wagon_files.append(fname)
Wagon_files, len(Wagon_files) 

Van_files =[]
for fname, lab in zip(labels['filename'], labels['label']):
  if lab =='Van':
    Van_files.append(fname)
Van_files, len(Van_files) 

Other_files =[]
for fname, lab in zip(labels['filename'], labels['label']):
  if lab =='Other':
    Other_files.append(fname)
Other_files, len(Other_files) 
```

```python
PATH = r'...\standford_cars\train'
Copy_to_path_sed = r'...\standford_cars\sedan_train\\'
Copy_to_path_con = r'...\standford_cars\convert_train\\'
Copy_to_path_suv = r'...\standford_cars\suv_train\\'
Copy_to_path_cou = r'...\standford_cars\cou_train\\'
Copy_to_path_cab = r'...\standford_cars\cab_train\\'
Copy_to_path_hat = r'...\standford_cars\hat_train\\'
Copy_to_path_min = r'...\standford_cars\min_train\\'
Copy_to_path_wag = r'...\standford_cars\wag_train\\'
Copy_to_path_van = r'...\standford_cars\van_train\\'
Copy_to_path_oth = r'...\standford_cars\oth_train\\'

for filename in os.listdir(PATH):
    img = Image.open(os.path.join(PATH, filename)) 
    if filename in Sedan_files:
        img.save(Copy_to_path_sed+filename) 
    if filename in Convertible_files:
        img.save(Copy_to_path_con+filename) 
    if filename in SUV_files:
        img.save(Copy_to_path_suv+filename) 
    if filename in Coupe_files:
        img.save(Copy_to_path_cou+filename)  
    if filename in Cab_files:
        img.save(Copy_to_path_cab+filename) 
    if filename in Hatchback_files:
        img.save(Copy_to_path_hat+filename) 
    if filename in Minivan_files:
        img.save(Copy_to_path_min+filename) 
    if filename in Wagon_files:
        img.save(Copy_to_path_wag+filename) 
    if filename in Van_files:
        img.save(Copy_to_path_van+filename) 
    if filename in Other_files:
        img.save(Copy_to_path_oth+filename) 
```

```python
from tensorflow.keras import layers
from tensorflow.keras import Model
import tensorflow as tf
import keras_preprocessing
from keras_preprocessing import image
from keras_preprocessing.image import ImageDataGenerator

TRAINING_DIR = r"...\standford_cars\train_data"
training_datagen = ImageDataGenerator(rescale = 1./255)

# VALIDATION_DIR = "/tmp/rps-test-set/"
# validation_datagen = ImageDataGenerator(rescale = 1./255)

train_generator = training_datagen.flow_from_directory(
        TRAINING_DIR,
        target_size=(100,100),
        class_mode='categorical',
        batch_size=16
)  
from tensorflow.keras.applications.inception_v3 import InceptionV3

local_weights_file = r'...\inception_v3_weights_tf_dim_ordering_tf_kernels_notop.h5'

pre_trained_model = InceptionV3(input_shape = (100, 100, 3), 
                                include_top = False,   # without fully connected layers
                                weights = None)   

pre_trained_model.load_weights(local_weights_file)

for layer in pre_trained_model.layers:
  layer.trainable = False
  
pre_trained_model.summary()

last_layer = pre_trained_model.get_layer('mixed9')   # the name of the choosing layer : mixed9
print('last layer output shape: ', last_layer.output_shape)
last_output = last_layer.output
```

```python
from tensorflow.keras.optimizers import RMSprop

# Flatten the output layer to 1 dimension using last_output as input:
x = layers.Flatten()(last_output)
# Add a fully connected layer with 1,024 hidden units and ReLU activation
x = layers.Dense(512, activation='relu')(x)
# Add a dropout rate of 0.2
x = layers.Dropout(0.2)(x)                  
# Add a final sigmoid layer for classification
x = layers.Dense(10, activation='softmax')(x)           

model = Model(pre_trained_model.input, x) 

model.compile(optimizer = RMSprop(lr=0.0001), 
              loss = 'binary_crossentropy', 
              metrics = ['accuracy'])
```

```python
history = model.fit(train_generator, steps_per_epoch = 20, epochs = 100, verbose = 1)
```

```python
# Loading the order of the image's name that has been provided
test_image_order = pd.read_csv(".../standford_cars/Testing_set.csv")
test_image_order.head()
```

```python
file_paths = [[fname, "...\standford_cars/test/" + fname] for fname in test_image_order['filename']]

test_images = pd.DataFrame(file_paths, columns=['filename', 'filepaths'])
test_images.head()
test_pixel_data = []     # initialize an empty numpy array
image_size = 100      # image size taken is 100 here. one can take other size too
for i in range(len(test_images)):
  img_array = cv2.imread(test_images['filepaths'][i])   # converting the image to gray scale
  new_img_array = cv2.resize(img_array, (image_size, image_size))      # resizing the image array
  test_pixel_data.append(new_img_array)

test_pixel_data = np.array(test_pixel_data)
test_pixel_data =  test_pixel_data.reshape(-1, 100, 100, 3)
pred = model.predict(test_pixel_data)
prediction = []
for value in pred:
  prediction.append(np.argmax(value))

# predictions = le.inverse_transform(prediction)
res = pd.DataFrame({'filename': test_images['filename'], 'label': prediction})  # prediction is nothing but the final predictions of your model on input features of your new unseen test data
res.to_csv("submission.csv", index = False)      # the csv file will be saved locally on the same location where this notebook is located.
```
